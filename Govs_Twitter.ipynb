```python
dados = pd.read_csv('/content/drive/MyDrive/VirtuÃÄ/Banco de dados/twitter 3.csv')
dados.head()
```

```python
list(dados['handle'].drop_duplicates())
```

```python
rodrigogarcia = (dados['handle'] == 'rodrigogarcia_')
sp = dados[rodrigogarcia].shape[0]
sp
```

```python
gov_sp = dados.loc[rodrigogarcia, ['handle', 'name', 'text']]
gov_sp.columns = ['usuario', 'nome', 'texto']
```

```python
gov_sp
```

```python
gov_sp.texto.str.split(expand=True).stack().value_counts()
```

```python
gov_sp['tweet'] = gov_sp['texto'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
gov_sp
```

```python
gov_sp.tweet.str.split(expand=True).stack().value_counts()
```

```python
gov_sp_lista = gov_sp.tweet.str.split(expand=True).stack().value_counts()
gov_sp_lista
```

```python
df_sp = pd.DataFrame(gov_sp_lista)
df_sp = df_sp.reset_index(level=0)
df_sp.columns = ['termos', 'total']
df_sp
```

```python
selecao_sp = df_sp['total'] >= 5
termos_sp = df_sp[selecao_sp]
termos_sp
```
